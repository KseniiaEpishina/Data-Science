### Data Preparation
--> Clean dataframe and select important info

```python

# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in df_melgyms.columns if col.startswith('location.')] + ['id']
df_melgyms_filtered = df_melgyms.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
df_melgyms_filtered['categories'] = df_melgyms_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
df_melgyms_filtered.columns = [column.split('.')[-1] for column in df_melgyms_filtered.columns]

df_melgyms_filtered.drop(['labeledLatLngs', 'formattedAddress'], axis=1, inplace=True)


df_melgyms_filtered
```

### Explore venue: Now Let's have a look at the existing venues around each gym in Sydney within a radius of 500 meters

```python

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['id'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Gym Name', 
                  'Gym Latitude', 
                  'Gym Longitude', 
                  'Venue',  
                  'Venue id',
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python

venues_around_melgyms = getNearbyVenues(names=df_melgyms_filtered['name'],
                                   latitudes=df_melgyms_filtered['lat'],
                                   longitudes=df_melgyms_filtered['lng']
                                  )

venues_around_melgyms.shape
```

### Cluster neighborhoods: preparing data for K-means clustering

###### Adding cluster labels to list of Gyms

```python
#Create clusters of Gyms based on the frequency of Venues by Venue Categories


# set number of clusters
kclusters = 4

melvenue_pivot_cat_clustering = melvenue_pivot_cat.drop(['Gym Name', 'Gym Latitude', 'Gym Longitude'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(melvenue_pivot_cat_clustering)

# check cluster labels generated for each row in the dataframe

kmeans.labels_[0:10]
```

```python
#gyms_merged = df_gyms_filtered
melgyms_merged=melvenue_pivot_cat

# add clustering labels
melgyms_merged['Cluster Labels'] = kmeans.labels_

melgyms_merged
cluster_summary=melgyms_merged.groupby(['Cluster Labels']).count()
cluster_summary['Gym Name']
```

```python
#Now lets look at the top 10 most 
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[3:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Gym Name']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
melgym_venues_sorted = pd.DataFrame(columns=columns)
melgym_venues_sorted['Gym Name'] = melgyms_merged['Gym Name']

for ind in np.arange(melgyms_merged.shape[0]):
    melgym_venues_sorted.iloc[ind, 1:] = return_most_common_venues(melgyms_merged.iloc[ind, :], num_top_venues)



melgym_venues_sorted
```
